# Contraste

```python
def change_contraste(image, factor):
    factor = float(factor)
    #element wise multiplication
    img1 = 128 + factor * image - factor * 128
    indx = np.where(img1<0)
    img1[indx] = 0
    indx = np.where(img1>255)
    img1[indx] = 255
    
    img1 = img1.astype(np.uint8)
    #img1 = np.clip(img1,0,255).astype(np.uint8)
    return img1
```

```python
img = plt.imread('../image/parrot.jpg')
```

```python
plt.imshow(img)
```

```python
img_con = change_contraste(img, factor = 1.5)
```

```python
plt.figure(figsize=(12,8))

plt.subplot(131)
plt.title("original")
plt.imshow(img)

plt.subplot(133)
plt.title("changed")
plt.imshow(img_con)
```
